```python
import seaborn as sns
df = sns.load_dataset('iris')
```

```python
df.shape
```

```python
df.info()
```

Draw Scatter Plot

```python
sns.regplot(x = df['sepal_length'],
            y = df['sepal_width']
).set(title = 'Tesa: Sepal length vs Sepal width',
      xlabel = 'Sepal length',
      ylabel = 'Sepal width')
```

```python
sns.regplot(x = df["petal_length"],
            y = df["petal_width"],
            scatter_kws = {"color": "red"}
).set(title = 'Tesa: Petal length vs Petal width',
      xlabel = 'Petal length',
      ylabel = 'Petal width')
```

```python
sns.set_style('dark')
sns.regplot(x = df['sepal_length'],
            y = df['sepal_width']
).set(title = 'Tesa: Sepal length vs Sepal width, Dark style',
      xlabel = 'Sepal length',
      ylabel = 'Sepal width')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 
```

```python
gapminder = pd.read_csv('https://raw.githubusercontent.com/holtzy/The-Python-Graph-Gallery/master/static/data/gapminderData.csv')
```

```python
gapminder.info()
```

```python
gapminder.continent.unique()
gapminder.year.unique()

data = gapminder.loc[gapminder.year == 2007]
```

```python
gapminder.tail(10)
```

```python
data.shape
```

```python
sns.set_theme(style = 'darkgrid')
sns.scatterplot(
    data = data,
    x = 'gdpPercap',
    y = 'lifeExp',
    size = 'pop',
    legend = False,
    sizes=(20,2000)
).set(title = 'Tesa: GDP per Capita vs Life expectancy',
      xlabel = 'GDP per Capita',
      ylabel = 'Life expectancy')

plt.show()
```

```python
sns.set_theme(style = 'darkgrid')
sns.scatterplot(
    data = data,
    x = 'gdpPercap',
    y = 'lifeExp',
    size = 'pop',
    legend = False,
    sizes=(20,200)
).set(title = 'Tesa: GDP per Capita vs Life expectancy',
      xlabel = 'GDP per Capita',
      ylabel = 'Life expectancy')

plt.show()
```

```python
sns.scatterplot(
    data = data,
    x = 'gdpPercap',
    y = 'lifeExp',
    size = 'pop',
    hue = 'country',
    legend = False,
    sizes=(20,2000)
).set(title = 'Tesa: GDP per Capita vs Life expectancy',
      xlabel = 'GDP per Capita',
      ylabel = 'Life expectancy')

plt.show()
```
